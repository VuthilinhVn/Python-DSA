# 算法引入

如果a+b+c = 1000, 且a^2+b^2 = c^2, 如何求出所有a,b,c可能的组合？

```python
import time

start_time = time.time()
for a in range(0, 1001):
    for b in range(0, 1001):
        for c in range(0, 1001):
            if (a+b+c)==1000 and a**2+b**2 == c**2:
                print(a, b, c)

end_time = time.time()
print(end_time-start_time)
print("done")


# 时间复杂度 T(n) = O(n^3)  因为有三个嵌套的循环，每个循环的迭代次数为 n（在这个例子中是 1000）。
```

# 时间复杂度 与 大O表示法

```python
# Solution 2:
import time

start_time = time.time()

for a in range(0, 1001):
    for b in range(0, 1001):
        c = 1000 - a - b
        if a**2+b**2 == c**2:
            print(a, b, c)

end_time = time.time()
print(end_time - start_time)
print("done")

# 时间复杂度 T(n) = O(n^2)  因为有两个嵌套的循环，每个循环的迭代次数为 n（在这个例子中是 1000）。
```
